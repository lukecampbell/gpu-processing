```python
nc_url = 'http://sos.maracoos.org/thredds/dodsC/agg/navhycom/agg.nc'
from netCDF4 import Dataset
nc = Dataset(nc_url, 'r')
nc.variables
u_var = nc.variables['water_u']
```

```python
v_var = nc.variables['water_v']
```

```python
u_var.dimensions
```

```python
lat = nc.variables['lat']
lon = nc.variables['lon']
```

```python
lat_data = lat[:]
lon_data = lon[:]
```

```python
i_0 = np.searchsorted(lat_data, [30,40])
```

```python
i_1 = np.searchsorted(lon_data, [290, 300])
```

```python
i_0
```

```python
i_1
```

```python
u_data = u_var[0, 0, i_0[0]:i_0[1], i_1[0]:i_1[1]]
```

```python
v_data = v_var[0, 0, i_0[0]:i_0[1], i_1[0]:i_1[1]]
```

```python
u_data.shape
```

```python
v_data.shape
```

```python
pylab.rcParams['figure.figsize'] = (20.0, 18.0)
from pylab import *
from numpy import ma

#1
fig = plt.subplot(111)
C = np.sqrt(u_data ** 2 + v_data ** 2)
Q = fig.quiver(u_data, v_data, C, cmap=plt.cm.jet)
```
